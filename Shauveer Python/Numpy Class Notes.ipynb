# List Comprehension

```python
# getting a list of fruits with letter 'a' in them

fruits = ['apple','banana','cherry','kiwi']
```

```python
# using for loop

a_fruits = []

for fruit in fruits:  # loop
    if 'a' in fruit:  # cond
        a_fruits.append(fruit)  # creation/append
        
print(a_fruits)
```

```python
# list comprehension

a1_fruits = [fruit for fruit in fruits if 'a' in fruit or fruit.startswith('c')]

print(a1_fruits)
```

## Arrays

Arrays: Sequence of adjacent memomy locations assinged to a single variable

- Complex Data types - built from simple data types eg int, char, float, boolean etc.
- int arrays - [1,2,3,4,5]
- char arrays - ['a','b','c']
- float arrays - [1.1,1.2,1.3,1.4]
- boolean arrays - [True,True,False]

> arr = [0,1,2,3,4,5,6]

Array indices start from 0 till N-1, N is the total number of elements in that array

Arrays are consuming less memory as compared to pyhton traditional list/tuple object

NumPy - python library to process such arrays/complex numeric types

ILATE - Inverse/Imaginary | Logarthimic | Algebric | Trignometric | Exponential

Matrix computations are also done in numpy

## Array Slicing

```python
arr = np.array([6,9,12,45,63,2,8,90,4,6,7,34,51])
print(arr[1])  # single number will refer to the index of the array elements
print(arr[1:4])  # x:y will refer to the sequence of indices from x index till y-1 index
print(arr[1:2])  # array output
print(arr[1:5:2])  # x:y:z will refer to every zth element from x:y slice
print(arr[4:9:3])
print(arr[:8:4])  # x is blank then from begining if y is blank then till end
```

# Case Study in Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageOps  # Python Image Library
```

```python
img = np.array(Image.open('landscape.jpg'))  # reading the image as numpy array
plt.figure(figsize=(8,8))
plt.imshow(img)
```

```python
img
```

```python
print(img.ndim)
print(img.shape)
```

```python
img_RG = img.copy()
```

```python
# Removing the Blue color from each pixel
img_RG[:,:,2] = 0
plt.figure(figsize=(8,8))
plt.imshow(img_RG)
```

```python
# removing the Green color from each pixel
img_RG[:,:,1] = 0
plt.figure(figsize=(8,8))
plt.imshow(img_RG)
```

```python
img_1 = img.copy()
```

```python
# Processing only a section of image
img_1[200:300,400:600,:2] = 255
plt.figure(figsize=(8,8))
plt.imshow(img_1)
```

```python
# Crop the image
img_2 = img.copy()
plt.figure(figsize=(8,8))
plt.imshow(img_2[:300,:500])
```

### Convert the image into B/W i.e. if the value of the color(R,G,B) > 128 then convert it into 255 
### if it is <=128 then convert it to zero

```python
img_BW = img.copy()

img_BW[img_BW[:,:,0] > 128] = 255  # converting Red intensity >128 to white
img_BW[img_BW[:,:,0] <= 128] = 0  # converting Red intensity <=128 to black

img_BW[img_BW[:,:,1] > 128] = 255  # converting Green intensity >128 to white
img_BW[img_BW[:,:,1] <= 128] = 0  # converting Green intensity <=128 to black

img_BW[img_BW[:,:,2] > 128] = 255  # converting Blue intensity >128 to white
img_BW[img_BW[:,:,2] <= 128] = 0  # converting Blue intensity <=128 to black

plt.figure(figsize=(8,8))
plt.imshow(img_BW)
```

# try and except blocks

```python
try:
    f = open('sample.txt')
except: # when try statement fails
    print('The file is no longer available')
else:  # when try statement is successful
    print('The file was found')
finally: # this will execute in all cases
    print('Thank you using our program')
```

```python
# advanced version of try and multiple except blocks

try:
    f = open('sample.txt')
except FilenotFound: # when try statement fails
    print('The file is no longer available')
except NoDBConnection: # when try statement fails
    print('The file is no longer available')
else:  # when try statement is successful
    print('The file was found')
finally: # this will execute in all cases
    print('Thank you using our program')
```
