```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### In-build Datasets in seaborn

```python
tips = sns.load_dataset('tips')
tips
```

# Regression Plot

This plot shows the line of regression (direction of relationship) between 2 numerical cols

```python
sns.regplot(x='total_bill',y='tip',data=tips)

# if it was suppose to be done using matplotlib
# plt.scatter()
# plt.xlabel
# ylabe
# plt.line(jbeg)
```

# Relational Plot
Scatter plot

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker')  # hue will create color filters
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')  # style will create markerstyle filters
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time',size='size')  # size will create small to big markersizes
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time',size='size',col='day')  # col will divide the plots into categories
```

# Heatmaps

visualize matrices as heat(color intensity) cells

```python
cormat = tips.corr()  # correlation matrix
cormat
```

```python
sns.heatmap(cormat,annot=True,fmt='.3f',cmap='viridis')
```

# Doubts

```python
# zip - paring the corresponding elements of 2 tuple objects
a = ('India','USA','China')
b = (3,5,7)
z = zip(a,b)
z = list(z)
print(z)
```

```python
# unzip(z)
first_element = [i for i,j in z]
second_element = [j+1 for i,j in z]
print(first_element)
print(second_element)
```
