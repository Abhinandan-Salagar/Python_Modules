# MatPlotLib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dayOfWeek = [1,2,3,4,5,6,7]
temp = [50,51,52,48,47,49,46]
```

```python
# Creating the most basic plot
plt.plot(dayOfWeek,temp)  # first arg will be X-axis | second will be Y-axis
# by default the plot will be a line chart
```

```python
# Playing with colors and linestyles
# linestyle symbols
plt.plot(dayOfWeek,temp,color='indigo',linestyle='-.' )
```

```python
#linestyle tuples
plt.plot(dayOfWeek,temp,color='lime',linestyle=(0, (3, 10, 1, 10)) )  # loosely dashdotted
```

```python
# adding visuals to the data points plotted known as markers
plt.plot(dayOfWeek,temp,color='olive',linestyle=':',marker="X",markersize=10)
```

```python
# specify color, linestyle and marker in a single argument
# generate a format string in the format -> [marker][line][color]
plt.plot(dayOfWeek,temp,'o--r')
```

```python
# Adding labels to the plot
plt.plot(dayOfWeek,temp,color='orchid',linestyle=(0, (3, 10, 1, 10)),marker="*")
plt.xlabel("Day of Week")  # give label to x-axis
plt.ylabel("Temperature")  # give label to y-axis
plt.title("Temperature vs Day of Week")  # add a title to the plot
```

```python
dayOfWeek = [1,2,3,4,5,6,7]
temp1 = [50,51,52,48,47,49,46]
temp2 = [43,42,40,44,33,35,37]
temp3 = [45,48,50,46,40,42,41]
```

```python
# adding multiple lines to the same plot
plt.plot(dayOfWeek,temp1,color='orchid',linestyle=(0, (3, 10, 1, 10)),marker="*",label='Week1')
plt.plot(dayOfWeek,temp2,color='olive',linestyle=(0, (1, 10)),marker="s",label='Week2')
plt.plot(dayOfWeek,temp3,color='orange',linestyle=(0, (5, 1)),marker="d",label='Week3')
plt.xlabel("Day of Week")  # give label to x-axis
plt.ylabel("Temperature")  # give label to y-axis
plt.title("Temperature vs Day of Week")  # add a title to the plot
plt.legend()  # will automatically take the styling from the plot and label arg to generate a legend 
```

# Bar plots
between 2 variables - 1 categorical and 1 is numerical column

```python
company = ['Realiance','ITC','Adani','Infosys','TCS']
revenue = [90,136,89,27,58]
```

```python
plt.bar(company,revenue)  # vertical bar plot
```

```python
plt.barh(company,revenue)  # horizontal bar plot
```

```python
plt.yticks(range(0,max(revenue) + 50,15))  # customize the value range of your plots
plt.bar(company,revenue,color='olive')
plt.title("Yearly Revenue")
plt.xlabel("Company Name")
plt.ylabel("Revenue(Billion$)")
```

# Histogram
The plot is made for a single numerical variable to showcase the distribution of values in class intervals

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]

# normal -> 80 - 100
# pre diabetic -> 100 - 125
# diabetic -> 125+
```

```python
plt.hist(blood_sugar,bins=[50,65,80,100,125,150])
```

```python
blood_pressure_men = np.random.randint(80,120,size=10)
print(blood_pressure_men)
blood_pressure_women = np.random.randint(80,120,size=10)
print(blood_pressure_women)
```

```python
plt.hist([blood_pressure_men,blood_pressure_women],bins=[80,85,100,110,125],color=['lavender','khaki'],label=['Men','Women'])
plt.legend()
plt.title("Blood Pressure - Women vs Men")
plt.xlabel("Blood Pressure Intervals")
plt.ylabel("Frequency")
```

# Create a histogram to represent the distribution of `Open`,`Close`,`High`,`Low` with proper labels, legends, titles on the AAPL.csv

```python
df = pd.read_csv('AAPL.csv')
df
plt.hist([df['Open'],df['Close'],df['High'],df['Low']],bins=(80,100,120,140,160,180,200),color=['lavender','khaki',"red","green"],label=["Open","Close","High","Low"])
plt.title('Histogram for prices')
plt.xlabel('Price Intervals')
plt.ylabel('Frequency')
plt.legend()
```

```python
import math
list(range(math.floor(df['Low'].min()),math.ceil(df['High'].max()),20))
```

```python
plt.hist([df['Open'],df['Close'],df['High'],df['Low']],bins=range(math.floor(df['Low'].min()),math.ceil(df['High'].max()),20),color=['red','green','blue','black'],label=['Open','Close','High','Low'])
plt.title('Histogram for prices')
plt.xlabel('Price Intervals')
plt.ylabel('Frequency')
plt.legend()
```

# Pie Charts
created for 2 variables, b/w categorical and numerical

```python
expenditures = [1400,600,300,410,250]
category = ["Rent","Food","Bills","Car","Others"]
```

```python
plt.pie(expenditures,labels = category, autopct='%0.0f%%',radius=1.5)
plt.plot()
```
