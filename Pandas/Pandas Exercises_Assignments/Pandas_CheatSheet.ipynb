```python
df1=pd.DataFrame({"id":[1,2,3,4,5,6],
                  "Grade":["A","B","C","D","E","F"],
                  "Marks":[10,20,30,40,50,60]})
df2=pd.DataFrame({"id":[1,2,3,4,5,6],
                  "Subject":["Math","Physics","Chemistry","Science","English","Hindi"],
                  "Name":["Abhi","Akash","Pawan","Vakya","Sona","Mona"]})
```

```python
pd.merge(df1,df2,how="inner",on="id")
```

```python
df1.merge(df2,how="left",on="id") # if you want to join two DataFrames based on a common column
```

```python
df2.merge(df1,how="right",left_on="id",right_on="id",right_index=False,left_index=False) 
```

```python
df2.join(df1,how="left",lsuffix="id") # if you want to join two DataFrames based on their indices, you can use the join() method
```

```python
df.groupby(df["EmployeeName"])
```

```python
# returns Series object that contains the count of non-null values 
df.groupby(["JobTitle"]).count()["BasePay"]
```

```python
df.groupby(["JobTitle"]).count()["BasePay"].values  # Return Series as ndarray
```

```python
df.groupby(["JobTitle"]).count()["BasePay"].value_counts() # Return a Series containing counts of unique values.
```

```python
df.groupby(["JobTitle"]).count()["BasePay"].reset_index() # reset_index will returns new df
```

```python
df.groupby(['Year']).mean()  # it gives entire df by grouping col
```

```python
# returns Series object that contains the count of non-null values
df.groupby(["Year"]).BasePay.mean()
```

```python
df.groupby(["Year"],as_index=False).BasePay.mean() # as_index will returns new df
```

```python
df.groupby(["Year"],dropna=True).sum()[["TotalPay","BasePay"]].reset_index()
```

```python
df.groupby(["Year","BasePay"]).filter(lambda x: x["Year"].iloc[0] > 2012).head(3)  # returns entire df by applying function on col
```

```python
df.groupby(["JobTitle"],as_index=False).BasePay.transform("count") # transform BasePay column to count of non-null BasePay values for each JobTitle
```

```python
df.groupby(["Year"])["BasePay"].mean().reset_index() # it returns mean value by grouping col
```

```python
df["Mean_BasePay"] = df.groupby(["Year"])["BasePay"].transform(np.mean) # it returns entire df with new col added for grouped mean value
df.head(5)
```

```python
df.groupby(["Year"])[["BasePay"]].agg(["min","max","count","std","mean","sum"]).reset_index()  # returns new df by all agg values
```

```python
df.groupby(["Year"]).agg({'BasePay':"mean",'TotalPay':'max','Benefits':"min"}) # returns new df by all agg values for each col
```
